```
#영화 공공데이터 : http://www.kobis.or.kr/kobisopenapi/homepg/apiservice/searchServiceInfo.do

url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.xml?key=430156241533f1d058c603178cc3ca0e&targetDt=20120101'
data = urlopen(url).read()
print(data)
```

```
#영화진흥 공공데이터 가져오기 : json형태 문자열로 가져오기 
def movieRank(cdate) :
    mykey = '430156241533f1d058c603178cc3ca0e'

    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json'
    url = url + '?key=' + mykey 
    url = url + '&targetDt=' + cdate

    data = urlopen(url).read().decode('utf-8')

    return data
```

```
#영화 json 문자열을 딕션너리로 변환 
#해당하는 영화 명을 리스트로 반환
def movieList(data) :
    mlt = []

    #딕션너리 변환
    dt = json.loads(data)

    #1위에서 10까지 영화 항목을 dl에 저장 : 1위 부터 10까지 딕션너리의 리스트 
    dl = dt['boxOfficeResult']['dailyBoxOfficeList'] 
    for item in dl: 
        mlt.append(item['movieNm']) 

    return mlt 
```

```
#리스트 항목보기
def show(mlt) :
    for item in mlt :
        print(item)
```

```
if __name__ == '__main__' :
    cdate = input('날짜입력(yyyymmdd)')
    data = movieRank(cdate) 
    mlist = movieList(data)

    show(mlist)
```
