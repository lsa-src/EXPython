해결문제) 부산쓰레기발생2017년.csv 파일에서 '금정구'의 연간 매립처리량과 소각처리량, 음식물류발생량 정보를 추출하시오.
+ 한글 인코딩 : euc-kr , cp949


```
연간(톤):매립처리량       7240.0
연간(톤):소각처리량        222.0
연간(톤):재활용처리량     56153.0
연간(톤):음식물류발생량    21273.0
```




````
import pandas as pd
````

````
with open('/content/drive/My Drive/Colab Notebooks/PythonBasic/Day07/부산쓰레기발생2017년.csv','r',encoding='cp949') as fp:
  data = fp.readlines()

lines = []

for line in data[1:] :
  lines.append(line)
print(lines)

dt = {'금정구':[]}
dt2 = {'항목':[]}

for line in lines :
  item = line.split(',')
  dt['금정구'].append(item[12])
  dt2['항목'].append(item[1] + ':' + item[0] ) 

print(dt)
print(dt2)
````

````
dsData = dt['금정구']
dsIndex = dt2['항목']
ds = pd.Series(dsData, index = dsIndex)

print(ds)
````

# 아래의 코드 활용을 잘 살펴보자

````
dtg = ds[[item for item in ds.index if item[:2] == '연간']]
dtg
````
