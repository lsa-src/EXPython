```
with open('/content/drive/My Drive/Colab Notebooks/PythonBasic/Day07/부산시기온.csv','r',encoding='utf-8') as fp:
  data = fp.readlines()
  dt = {'일시':[], '평균':[], '최고':[], '최저':[]}

for line in data :
  line = line.replace('\n','')

for line in data[1: ]:
  #for i in range(1, len(data)) :
  #if line[:2] =='일시' : continue
  item = line.split(',')
  dt['일시'].append(item[0])
  dt['평균'].append(float(item[1]))
  dt['최고'].append(float(item[2]))
  dt['최저'].append(float(item[3]))

print(dt['일시'])
print(dt['평균'])
print(dt['최고'])
print(dt['최저'])

print(type(dt))
print(dt)
```

##Series
+ index로 value를 구분하는  1차원 자료구조

```
tempavg = list(dt['평균'])
print(tempavg)
print(type(tempavg))
```

+ **리스트(list)를 판다스 시리즈(series)로 만들기**

### series 기초 통계
+ mean() : 평균
+ max() : 최대
+ min() : 최소
+ std() : 표준편차
+ describe() : 요약 통계량 계산

```
print(f'평균 : {ds.mean():0.2f}')
print(f'최대 : {ds.max():0.2f}')
print(f'최소 : {ds.min():0.2f}')
print(f'표준편차 : {ds.std():0.2f}')
print(f'요약 \n{ds.describe() }')
```

```
print(f'{ds.mean():0.2f}')
dsup1 = ds[ds > 14]
print(dsup1)
dsup2 = ds[ds > ds.mean()] #시리즈에서 평균보다 큰 값(행)을 찾아라
print(dsup2)
```

### series 자주사용되는 메소드
+ replace() : 특정값을 가진 요소 값을 변경
+ equals() : 시리즈와 시리즈가 같은지 비교
+ isin() : 시리즈에 포함된 값이 있는지 확인 
+ drop_duplicates() : 중복값이 없는 시리즈 반환
+ append() : 2개이상의 시리즈 연결 
+ sort_index() : 인덱스로 정렬
 - ascending = False : 역순 정렬 옵션
+ sort_values() : 값으로 정렬
+ to_frame() : 시리즈를 데이터프레임으로 변경

```
ds0 = ds #복사
print(ds0)
```

+ replace() : 특정값을 가진 요소 값을 변경

```
#값 변경 : 21.3인 값을 21.5로 변경
ds0[ds0 == 21.3]
```

```
ds0 = ds0.replace(21.3, 21.5)
ds0['2017-06-01']
```

+ equals() : 시리즈와 시리즈가 같은지 비교

```
#두 시리즈 같은지 비교 객체비교, 개수부터 확인
print(ds0.equals(ds))
```

+ isin() : 시리즈에 포함된 값이 있는지 확인

```
#값에 리스트 내용이 있는지 확인 
ds0.isin([21.5])
```

```
#값에 리스트 내용이 있는 항목의 개수 
ds0.isin([21.5]).sum()

ds215 = ds0.isin([21.5])
type(ds215) #키, 값(T F)
ds215.sum()
```

+ drop_duplicates() : 중복값이 없는 시리즈 반환

```
#값이 동일한 
ds0 = ds0.drop_duplicates()
ds0.describe() #2개 날라감. count -2 됨
```

+ append() : 2개이상의 시리즈 연결 

```
ds1 = ds[:'2017-06-01']
ds2 = ds['2018-01-01':'2018-06-01']
print(ds1)
print(ds2)
```

```
ds21 = ds2.append(ds1)
ds21
```

+ sort_index() : 인덱스로 정렬
 - ascending = False : 역순 정렬 옵션

```
ds21 = ds21.sort_index() #key순으로 정렬
ds21
```

```
ds21 = ds21.sort_index(ascending = False)
ds21
```

+ sort_values() : 값으로 정렬

```
ds21 = ds21.sort_values()
ds21
```

+ to_frame() : 시리즈를 데이터프레임으로 변경


```
ds21 = ds21.to_frame()
print(type(ds21))
```

### series 연산 (인덱스가 같은 것끼리 연산)
+ \+ : 같은 길이의 시리즈와 시리즈 더하기
+ \* : 같은 길이의 시리즈와 시리즈 곱하기


```
print(ds1 + 1)
print(ds2 * 3)
```

```
ds1 + ds2 #키값이 달라서 연산 안됨
```

```
with open('/content/drive/My Drive/PythonWork/PythonBasic/04/부산쓰레기발생2017년.csv','r',encoding='cp949') as fp:
    data = fp.readlines() 

dt = {'인덱스':[], '금정구':[]}
for line in data:
    if line[:2] == '항목' : continue

    line = line.replace('\n','')
    
    item = line.split(',')
    dt['인덱스'].append(item[1] + ':' + item[0] ) 
    dt['금정구'].append(float(item[12])) 
```

```
dtgubun = pd.Series(dt['금정구'], index=dt['인덱스'])
dtgubun
```
