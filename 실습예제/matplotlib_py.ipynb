```python
plt.figure(figsize=(10,5)) #가로세로 튜플

plt.subplot(1,3,1)
plt.plot(temp, 'go-.') 

plt.subplot(1,3,2)
plt.plot(temp, color ='r', marker ='+') 

plt.subplot(1,3,3)
plt.plot(temp, color ='b', marker ='x')

plt.show()
```

```python
fig =  plt.figure(figsize=(10,5)) #가로세로 튜플

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

ax1.plot(temp, 'go--')
ax1.set_title('green')
ax2.plot(temp, 'rx--')
ax2.set_title('red')
ax3.plot(temp, 'b^--')
ax3.set_xlabel('blue')
```

```python
#2019년 부산의 1월에서 12월 평균, 최저, 최고 기온
avg = [4.5,6.6,10.5,13.4,19.3,21.1,24.3,27,23.1,18.6,13,7]
min = [-4.4,-4.2,0.5,2.1,11,15.5,19.8,20.3,15.8,10.4,2,-2.8]
max = [13.7,16.5,21,23.4,30.1,29.7,30.9,35,30.5,28.8,24,16.6]
```

```python
plt.figure(figsize=(10,5)) #가로세로 튜플
'''
x=[]
for i in range(len(avg)) :
    x.append(str(i+1)+'월')
plt.bar(x, avg)
'''
x = list(range(len(avg)))
xm= [str(i+1)+'월' for i in x]

#마이너스 기호 표시
plt.rcParams['axes.unicode_minus'] = False
plt.bar(xm, avg, label='평균')
plt.plot(min, 'bo-.', drawstyle='steps-post', label='최저')
plt.plot(max, 'ro-.', label='최고')
plt.title("3월 기온변화")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5)) #가로세로 튜플
'''
x=[]
for i in range(len(avg)) :
    x.append(str(i+1)+'월')
plt.bar(x, avg)
'''
x = list(range(len(avg)))
xm= [str(i+1)+'월' for i in x]

#마이너스 기호 표시
plt.rcParams['axes.unicode_minus'] = False

plt.bar(xm, max, label='최고')
plt.bar(xm, avg, label='평균')
plt.bar(xm, min, label='최저')

plt.title("3월 기온변화")
plt.legend()
plt.show()
```
