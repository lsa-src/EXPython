```
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
```

### 2. Url 추출



```
def show(cDate ) :
  url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.xml?key=430156241533f1d058c603178cc3ca0e'
  url = url +'&targetDt='+cDate

  memoData = urlopen(url) # 가져온 객체를 decode
  content = bs(memoData, 'lxml-xml') 
  print(type(content))
  return content
```

```
cDate=input('1. 일간 박스 오피스 (날짜 입력 : yyyymmdd ) : ')
data1=show(cDate)
```

```
items = data1.find_all('dailyBoxOffice')

rank = []
movieNm = []

for item in items:
    rank.append(item.find('rank').text)
    movieNm.append(item.find('movieNm').text)
  
for item in rank,  movieNm :
    print(item)
```
