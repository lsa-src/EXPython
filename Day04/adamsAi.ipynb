```
from urllib.request import urlopen
from urllib import parse

import json 
```

```
def adamsAi(iTxt,  myType) :
  mykey='4722875241399063807'
  myquery=parse.quote(iTxt)

  url = 'http://api.adams.ai/datamixiApi/omAnalysis?key='+ mykey
  url =  url +'&query='+myquery
  url =  url +'&type='+ myType
  print(url)

  dt = urlopen(url).read().decode('utf-8')
  dt = json.loads(dt)
  # print(dt)
  # print(type(dt))
  rt = dt['return_object']
  score  = rt['score']
  label = rt['label']
  return score, label
```

```
iTxt = input('분석문장을 넣어주세요 : ')
myType = input('[0:감성분석, 1: 감정분석]')

score, label = adamsAi(iTxt, myType)
print ( iTxt + '의 감성분석 결과 : ' + str(score), label ) 
```
