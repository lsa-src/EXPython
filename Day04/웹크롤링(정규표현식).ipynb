```
s = '별점 - 총 10점 중10	OO 브라다의 의리 멋있다! 황정민 대박 	silv****	20.3.6'
pat =r'([a-z]+)([*]+)' #작성자 뽑기, ( ) - 그루핑
writer = re.search(pat, s)

print( writer.group() )
print( writer.group(1) )
print( writer.group(2) )
```

```
s = '별점 - 총 10점 중10	OO 브라다의 의리 멋있다! 황정민 대박 	silv****	20.3.6'
pat =r'([\d]+)[.]([\d]+)[.]([\d]+)' #년월일- 그루핑
writer = re.search(pat, s)

print( writer.group() )
print( writer.group(1), '년', writer.group(2), '월', writer.group(3), '일' ) #년 월 일
```

```
s = '별점 - 총 10점 중10	OO 브라다의 의리 멋있다! 황정민 대박 	silv****	20.3.6'
pat =r'([a-zA-Z0-9])+([*]+)' #작성자
writer = re.search(pat, s).group()


pat =r'[중]([\d]+)' #평점 
jum = re.search(pat, s).group(1)

print(writer,":", jum,"점" )

```

```
def writerScore(s):
  lt=[]
  pat = r'([a-zA-Z0-9]+)([*]+)' #작성자
  lt.append(re.search(pat,s).group(1))

  pat =r'[중]([\d]+)' #평점 
  lt.append(re.search(pat, s).group(1))

  return lt
```

```
with open('/content/drive/My Drive/Colab Notebooks/PythonWork/PythonBasic/work/Day04/평점.txt', 'r', encoding ='utf-8') as fp:
  data = fp.readlines()

dt={}

for item in data :
  lt =  writerScore(item)
  dt[lt[0]] = int(lt[1])

print(dt)
  
```

```
x = list(dt.keys())
y = list(dt.values())
avg = sum(y)/len(y)
print('평균 : ', avg )
print(x, y)

import matplotlib.pyplot as plt

plt.bar(x, y)
plt.axhline(avg, color='r', linewidth=1)

plt.show()
```
